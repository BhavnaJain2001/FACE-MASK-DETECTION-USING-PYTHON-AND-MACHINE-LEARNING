```python
import numpy as np
import cv2
```

```python
with_mask = np.load('with_mask.npy')
without_mask = np.load('without_mask.npy')
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
with_mask = with_mask.reshape(200,50 * 50 *3)
without_mask = without_mask.reshape(200,50 * 50 *3)
```

```python
with_mask.shape
```

```python
without_mask.shape
```

```python
X = np.r_[with_mask, without_mask]
```

```python
X.shape
```

```python
labels = np.zeros(X.shape[0])
```

```python
labels[200:] = 1.0
```

```python
names = {0 : 'Mask', 1 : 'No Mask'}
```

# svm = Support Vector Machine
# svc = Support Vector Classification

```python
from sklearn.svm  import SVC
from sklearn.metrics import accuracy_score
```

```python
from sklearn .model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(X,labels, test_size=0.25)
```

```python
x_train.shape
```

# PCA = Principal Component Analysis

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components = 3)
x_train = pca.fit_transform(x_train)
```

```python
x_train[0]
```

```python
x_train.shape
```

```python
svm = SVC()
svm.fit(x_train, y_train)
```

```python
import pickle

with open('pca_model.pkl', 'wb') as pca_file:
    pickle.dump(pca, pca_file)

with open('svm_model.pkl', 'wb') as svm_file:
    pickle.dump(svm, svm_file)
```

```python
x_test = pca.transform(x_test)
y_pred = svm.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
from flask import Flask, render_template, Response
import cv2
import numpy as np
import pickle

app = Flask(__name__)

faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + 'haarcascade_frontalface_default.xml')
capture = cv2.VideoCapture(0)

# Load your PCA and SVM models here
pca = pickle.load(open('pca_model.pkl', 'rb'))
svm = pickle.load(open('svm_model.pkl', 'rb'))
names = {0 : 'Mask', 1 : 'No Mask'}

font = cv2.FONT_HERSHEY_COMPLEX

def generate_frames():
    while True:
        flag, img = capture.read()
        if flag:
            faces = faceCascade.detectMultiScale(img, scaleFactor=1.1, minNeighbors=5, minSize=(30, 30))
            for (x, y, w, h) in faces:
                cv2.rectangle(img, (x, y), (x+w, y+h), (255, 0, 255), 4)
                face = img[y:y+h, x:x+w]
                face = cv2.resize(face, (50, 50))
                face = face.reshape(1, -1)
                face = pca.transform(face)
                pred = svm.predict(face)
                n = names[int(pred)]
                cv2.putText(img, n, (x, y), font, 1, (244, 250, 250), 2)
            
            ret, buffer = cv2.imencode('.png', img)
            frame = buffer.tobytes()
            yield (b'--frame\r\n'
                   b'Content-Type: image/png\r\n\r\n' + frame + b'\r\n')
        else:
            break

@app.route('/')
def index():
    return render_template('index.html')

@app.route('/video_feed')
def video_feed():
    return Response(generate_frames(), mimetype='multipart/x-mixed-replace; boundary=frame')

if __name__ == '__main__':
    app.run(debug=True)
```
